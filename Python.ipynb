```python
# Hotel Management System with Room IDs and Payment System

# Step 1: Define Rooms
room_types = ["Single", "Double", "Deluxe", "Suite"]
room_prices = [1500, 2500, 4000, 6000]
room_availability = [5, 3, 2, 1]

# Step 2: Display Available Rooms
print("Available Rooms:")
print("ID\tType\tPrice/Night\tAvailable")
for i in range(len(room_types)):
    print(f"{i+1}\t{room_types[i]}\tâ‚¹{room_prices[i]}\t\t{room_availability[i]}")

# Lists to store booking details
booked_rooms = []
booked_nights = []
booked_prices = []

# Step 3: Function to Book a Room
def book_room(room_id, nights):
    index = room_id - 1
    room_name = room_types[index]
    price_per_night = room_prices[index]

  # Check room availability
    if room_availability[index] <= 0:
        print(f" Sorry, {room_name} rooms are fully booked.")
        return

    if room_name in booked_rooms:
        pos = booked_rooms.index(room_name)
        booked_nights[pos] += nights
        booked_prices[pos] += price_per_night * nights
        print(f"Updated: {room_name} total nights = {booked_nights[pos]}, Total = â‚¹{booked_prices[pos]}")
    else:
      room_availability[index] -= 1

      total_price = price_per_night * nights

    # Add to booking list
      booked_rooms.append(room_name)
      booked_nights.append(nights)
      booked_prices.append(total_price)


      print(f"{room_name} booked for {nights} night(s). Total = â‚¹{total_price}")
# Step 4: Main Program
while True:
    choice = input("\nEnter Room ID to book (or 'done' to checkout): ").strip()

    if choice.lower() == "done":
        break

    if not choice.isdigit():
        print("âš  Please enter a valid Room ID (1â€“4)!")
        continue

    room_id = int(choice)

    if room_id < 1 or room_id > len(room_types):
        print("âš  Invalid Room ID! Please choose between 1 and 4.")
        continue

    nights_input = input("Enter number of nights: ").strip()
    if not nights_input.isdigit():
        print("âš  Please enter a valid number for nights!")
        continue

    nights = int(nights_input)

    book_room(room_id, nights)


# Step 5: Payment System
def payment(grand_total):
    print("\n Payment Options:")
    print("1. Cash")
    print("2. Card")
    print("3. UPI")

    option = input("Choose your payment method (1/2/3): ")

    if option == "1":
        print("\n You selected: Cash")
        amount = input("Enter amount given by customer: ")
        if amount.isdigit():
            amount = int(amount)
            if amount >= grand_total:
                balance = amount - grand_total
                print(f" Payment successful! Change to return: â‚¹{balance}")
            else:
                print(" Insufficient cash amount. Payment failed.")
                return False
        else:
            print(" Invalid amount entered.")
            return False

    elif option == "2":
        print("\n You selected: Card Payment")
        card_no = input("Enter last 4 digits of card: ")
        if len(card_no) == 4 and card_no.isdigit():
            print(" Card payment successful!")
        else:
            print(" Invalid card number.")
            return False

    elif option == "3":
        print("\nðŸ“± You selected: UPI Payment")
        upi_id = input("Enter UPI ID: ")
        if "@" in upi_id:
            print(" UPI payment successful!")
        else:
            print(" Invalid UPI ID.")
            return False

    else:
        print("Invalid payment option.")
        return False

    print("-------------------------------------------------------------")
    print("Payment Receipt:")
    print(f"Total Paid: â‚¹{grand_total}")
    print("Payment Status: Successful")
    print("-------------------------------------------------------------")
    return True

# Step 6: Payment Summary
print("\n ====== BOOKING SUMMARY ======")
print("Room Type\tNights\tTotal (â‚¹)")
total_amount = 0

for i in range(len(booked_rooms)):
    print(f"{booked_rooms[i]}\t\t{booked_nights[i]}\tâ‚¹{booked_prices[i]}")
    total_amount += booked_prices[i]

print("-----------------------------------")
print(f" Total Payable Amount: â‚¹{total_amount}")
print(" Thank you for booking with our hotel!")
```

```python
# Online Food Delivery System with Menu IDs and Payment System

# Step 1: Define Menu
food_items = ["Burger", "Pizza", "Pasta", "Maggi"]
food_prices = [120, 250, 180, 150]
food_availability = [10, 8, 6, 5]

print("Available Food Items:")
print("ID\tItem\tPrice\tAvailable")

for i in range(len(food_items)):
    print(f"{i+1}\t{food_items[i]}\tâ‚¹{food_prices[i]}\t\t{food_availability[i]}")

cart_items = []
cart_qty = []
cart_total = []

def order_food(food_id, qty):
    index = food_id - 1

    if food_availability[index] < qty:
        print("âŒ Not enough stock available")
        return

    food_availability[index] -= qty
    total = food_prices[index] * qty

    cart_items.append(food_items[index])
    cart_qty.append(qty)
    cart_total.append(total)

    print(f"{food_items[index]} added | Qty: {qty} | Total: â‚¹{total}")

while True:
    choice = input("\nEnter Food ID (or 'done'): ")
    if choice.lower() == "done":
        break

    qty = int(input("Enter Quantity: "))
    order_food(int(choice), qty)

print("\n ====== ORDER SUMMARY ======")
grand_total = sum(cart_total)
for i in range(len(cart_items)):
    print(f"{cart_items[i]} x {cart_qty[i]} = â‚¹{cart_total[i]}")

print(f"Total Payable Amount: â‚¹{grand_total}")

```

```python
# Railway Reservation System with Train IDs and Ticket Booking System

# Step 1: Train Details
train_names = ["Express", "sleeper", "Local"]
train_fare = [800, 1200, 300]
train_seats = [5, 3, 10]

print(" Available Trains")
print("ID\tTrain\tFare\tSeats")
for i in range(len(train_names)):
    print(f"{i+1}\t{train_names[i]}\tâ‚¹{train_fare[i]}\t{train_seats[i]}")

booked_train = []
booked_seats = []
booked_amount = []

def book_ticket(train_id, seats):
    index = train_id - 1

    if train_seats[index] < seats:
        print(" Not enough seats available")
        return

    train_seats[index] -= seats
    total = train_fare[index] * seats

    booked_train.append(train_names[index])
    booked_seats.append(seats)
    booked_amount.append(total)

    print(f" {seats} seat(s) booked in {train_names[index]} | â‚¹{total}")

while True:
    choice = input("\nEnter Train ID (or 'done'): ")
    if choice.lower() == "done":
        break

    seats = int(input("Enter number of seats: "))
    book_ticket(int(choice), seats)

print("\n ====== BOOKING SUMMARY ======")
total_pay = sum(booked_amount)
for i in range(len(booked_train)):
    print(f"{booked_train[i]} - Seats: {booked_seats[i]} - â‚¹{booked_amount[i]}")

print(f"Total Fare: â‚¹{total_pay}")
```

```python
# Restaurant Ordering System with Item IDs and Billing System

# Step 1: Restaurant Menu
menu_items = ["Dosa", "Samosa", "Bhel", "Noodles"]
menu_price = [100, 80, 900, 180]

ordered_items = []
ordered_qty = []
ordered_price = []

print("ðŸ´ Restaurant Menu")
print("ID\tItem\tPrice")
for i in range(len(menu_items)):
    print(f"{i+1}\t{menu_items[i]}\tâ‚¹{menu_price[i]}")

def place_order(item_id, qty):
    index = item_id - 1
    total = menu_price[index] * qty

    ordered_items.append(menu_items[index])
    ordered_qty.append(qty)
    ordered_price.append(total)

    print(f" {menu_items[index]} ordered | Qty: {qty} | â‚¹{total}")

while True:
    choice = input("\nEnter Item ID (or 'done'): ")
    if choice.lower() == "done":
        break

    qty = int(input("Enter Quantity: "))
    place_order(int(choice), qty)

print("\n BILL SUMMARY")
bill = sum(ordered_price)
for i in range(len(ordered_items)):
    print(f"{ordered_items[i]} x {ordered_qty[i]} = â‚¹{ordered_price[i]}")

print("====== Order SUMMARY ======")
print(f"Total Bill Amount: â‚¹{bill}")
```

```python
# ATM Management System with PIN Authentication and Transaction System


balance = 5000
pin = "1234"

def atm_menu():
    print("\n ATM MENU")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

entered_pin = input("Enter ATM PIN: ")

if entered_pin == pin:
    while True:
        atm_menu()
        choice = input("Choose option: ")

        if choice == "1":
            print(f"Balance: â‚¹{balance}")

        elif choice == "2":
            amount = int(input("Enter deposit amount: "))
            balance += amount
            print("Amount Deposited")

        elif choice == "3":
            amount = int(input("Enter withdraw amount: "))
            if amount <= balance:
                balance -= amount
                print("Please collect cash")
            else:
                print("Insufficient balance")

        elif choice == "4":
            print("Thank you for using ATM")
            break

        else:
            print("Invalid choice")
else:
    print("Incorrect PIN")
```
